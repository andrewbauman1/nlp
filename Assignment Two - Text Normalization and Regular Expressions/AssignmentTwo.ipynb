**5.Find instances of the sequences of letters  `wood` and `chuck` in the text below**<br>
"How much wood would a woodchuck chuck if a woodchuck could chuck wood"


```python
# Enter your code here:
txt = "How much wood would a woodchuck chuck if a woodchuck could chuck wood"
match = re.search("wood|chuck", txt)

print(re.findall("wood|chuck", txt))
for match in re.finditer("wood|chuck", txt):
  print(match.span())
```

**6. Using the text below, extract the extension 3598** <br>
"Joe Smith, Sr. VP, x3598, jsmith@greatcompany.com" 

```python
# Enter your code here:
txt2 = "Joe Smith, Sr. VP, x3598, jsmith@greatcompany.com"
r = re.compile(r'x\d{4}')
ext = r.search(txt2)[0]
print(ext)
```

**7. Using the text below, extract any email address contained in it.** <br>
"Joe Smith, Sr. VP, x3598, jsmith@greatcompany.com" 

```python
# Enter your code here:
emails = re.findall(r'\b[\w.]+@+[\w.]+.+[\w.]\b', txt2)
for email in emails:
  print(email)
```

**8. Using the sequence of characters below, split it into a list of numbers using the letters as separators .** <br>
"17a5bb3c45g789f" 

```python
# Enter your code here
rawList = "17a5bb3c45g789f"
numberList = re.findall("\d+", rawList)

print(numberList)
```
